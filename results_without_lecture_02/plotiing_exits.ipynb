```python
import pandas as pd
import numpy as np
import glob
import itertools
import os
import matplotlib.pyplot as plt 
```

```python
general_directory = os.getcwd()
folder_name = general_directory.split('/')[-1]
```

```python
first_directory = os.path.join(general_directory,"flow_exit_id_2_trajectories_scenario_01.txt")
directories = []
ids = [1, 2, 5, 6, 7, 8, 9, 11, 12, 13,14, 15,16,19,20,22,23,24,25,26,27,28,29]
for id in ids:
    name = "flow_exit_id_"+ str(id) +"_trajectories_scenario_01.txt"
    directories.append(os.path.join(general_directory,name))
```

```python
def reading_data(directory):
    col_names = ["Time", "cum_agents", "id"]
    reading_file=pd.read_csv(directory,delimiter=' ',skiprows=3,names=col_names)
    x=reading_file["Time"]
    y=reading_file["cum_agents"]
    return x, y
```

```python
marker = itertools.cycle(('o','^','+','s','p'))
color = itertools.cycle(('r','g','b','c','m','y','k'))
linestyle=itertools.cycle(('--','-'))
id_iterators = itertools.cycle(ids)
def data_plot(x,y):
    plt.plot(x,y,linestyle=next(linestyle),color=next(color),label=next(id_iterators))
    plt.ylabel("cumulative agents")    
    plt.xlabel("Time")
```

```python
for directory in directories:
    x,y=reading_data(directory)
    data_plot(x,y)

plt.legend();
name = folder_name + ('.png')
plt.savefig(name)
```
